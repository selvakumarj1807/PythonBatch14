# Tuple
--------

1. Tuple is a collection of Data type or Built in data types
2. It can declare - ( )
3. It can store n number of Values and Diffrent Data Type Values in a single Variable. Based on Index
4. Tuple is a Immutable(UnChangeable)

```python
userData = (101, 'Sara', 25, 'Female')

print(userData, type(userData))
```

# Type Casting

```python
a = 10

b = 10.54321

print(a, type(a))

print(b, type(b))

print('--------------------------')

a = float(a)

b = int(b)

print(a, type(a))

print(b, type(b))
```

```python
userData = (101, 'Sara', 25, 'Female')

print(userData, type(userData))

print('---------------------------------------')

print(userData[2])

userData = list(userData)

print(userData, type(userData))

userData[2] = 30

print(userData[2])

print('---------------------------------------')

userData = tuple(userData)

print(userData, type(userData))
```
