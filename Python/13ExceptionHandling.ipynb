# Python Exception Handling or Error Handling

# Try & Except

```python
try:
    a = 10

    b = 0
    
    c = a / b
    
    print(c)

except:
    print('Divisor Cannot be Zero!...')
```

# Exception

```python
try:
    a = 10

    b = 0
    
    c = a / b
    
    print(c)

except Exception as e:
    print('Divisor Cannot be Zero!...')

    print('Try block have Some of Errors : ', e)
```

# Try & Except & Else

```python
try:
    a = 10

    b = 0
    
    c = a / b

except Exception as e:
    print('Try block have Some of Errors : ', e)

else:
    print(c)
```

# Try & Except & Else & Finally

```python
try:
    a = 10

    b = 2
    
    c = a / b

except Exception as e:
    print('Try block have Some of Errors : ', e)

else:
    print(c)

finally:
    print('Thank You Come Again!....')
```

# ValueError

```python
try:
    name = input('Enter your Name : ')

    age = int(input('Enter your Age : '))

    print(name, age)

except Exception as e:
    print('Age Values Must be Declare Integer Value!..')

    age = int(input('Enter your Age : '))

    print(name, age)
```
