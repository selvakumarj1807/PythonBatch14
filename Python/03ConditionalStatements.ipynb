# If

```python
age = 22

print(age >= 18)

if age >= 18:
    print(age, 'above 18. So You are eligible for Vote')
```

# If & Else

```python
age = 14

print(age >= 18)

if age >= 18:
    print(age, 'above 18. So You are eligible for Vote')
else:
    print(age, 'below 18. So You are Not eligible for Vote')
```

# If & Elif & Else

```python
average = int(input('Enter your Average value : '))

if average >= 90 and average <= 100:  
    print('A Grade!...')
elif average >= 80 and average <= 90:
    print('B Grade!...')
elif average >= 70 and average <= 80:
    print('C Grade!...')
else:
    print('No Grade!...')
```

# Nested If

```python
m1 = int(input('Enter Mark 01 : '))

m2 = int(input('Enter Mark 02 : '))

m3 = int(input('Enter Mark 03 : '))

print('')

if m1 >= 35 and m2 >= 35 and m3 >= 35:
    print('You are Pass!...')
    total = m1 + m2 + m3
    average = total / 3

    print('Total : ', total)
    print('Average : ', int(average))

    if average >= 90 and average <= 100:
        print('A Grade!...')
    elif average >= 80 and average <= 90:
        print('B Grade!...')
    elif average >= 70 and average <= 80:
        print('C Grade!...')
    else:
        print('No Grade!...')
else:
    print('You are Fail!...')

    if m1 <= 35:
        print(m1, '- Mark 01 is Below 35')

    if m2 <= 35:
        print(m2, '- Mark 02 is Below 35')

    if m3 <= 35:
        print(m3, '- Mark 03 is Below 35')
```
