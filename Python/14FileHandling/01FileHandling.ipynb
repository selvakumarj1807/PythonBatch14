# File Read

```python
readFile = open('sample.txt', 'r')

print(readFile.read())

readFile.close()
```

# File Write

```python
readFile = open('sample.txt', 'r')

print(readFile.read())

readFile.close()

print('--------------------------------------')

writeFile = open('sample.txt', 'w')

writeFile.write('This is File Handling Write Method!...')

writeFile.close()
```

# File Append

```python
readFile = open('sample.txt', 'r')

print(readFile.read())

readFile.close()

print('--------------------------------------')

writeFile = open('sample.txt', 'a')

writeFile.write(' This is File Handling Write Method!...')

writeFile.close()
```
