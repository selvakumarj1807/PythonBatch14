# 02 No Return Type With Arguments

```python
# Create the Function

def addtwoNumber(a, b):
    c = a + b
    print(a, '+', b, '=', c)
```

```python
# Call the Function

addtwoNumber(10, 2)

print('----------------------------------')

addtwoNumber(20, 4)
```

# 03 Return Type With Out Arguments

```python
# Create the Function

def addtwoNumber():
    a = int(input('Enter your Value 01 : '))
    b = int(input('Enter your Value 02 : '))

    c = a + b
    return c
```

```python
# Call the Function

print('Get Print Statement : ', addtwoNumber())

print('----------------------------------')

result01 = addtwoNumber()

print('Get Variable : ', result01)
```

# Return Type With Arguments

```python
# Create the Function

def addtwoNumber(a, b):
    c = a + b
    return c
```

```python
# Call the Function

print('Get Print Statement : ', addtwoNumber(10, 2))

print('----------------------------------')

result01 = addtwoNumber(20, 4)

print('Get Variable : ', result01)
```

```python
# Create the Function

def addtwoNumber(a, b):
    c = a + b
    return f'{a} + {b} = {c}'
```

```python
# Call the Function

print('Get Print Statement : ', addtwoNumber(10, 2))

print('----------------------------------')

result01 = addtwoNumber(20, 4)

print('Get Variable : ', result01)
```
