# Decorator Property

# Property Getter

```python
class Student:
    def __init__(self, total):
        self._total = total

    @property
    def total(self):
        return self._total
```

```python
studentObject01 = Student(250)

print(studentObject01.total)
```

# Property Setter

```python
class Student:
    def __init__(self, total):
        self._total = total

    @property
    def total(self):
        return self._total

    @total.setter
    def total(self, t):
        print('Setter : ', t)
```

```python
studentObject01 = Student(450)

print(studentObject01.total)

studentObject01.total = 500
```

# Property Getter & Property Setter

```python
class Student:
    def __init__(self, total):
        self._total = total

    @property
    def total(self):
        return f"Total : {self._total}"

    @total.setter
    def total(self, t):
        if t <= 0 or t > 500:
            print('Give a Valid Total')
        else:
            self._total = t
```

```python
studentObject01 = Student(0)

print(studentObject01.total)

studentObject01.total = 450

print(studentObject01.total)
```
