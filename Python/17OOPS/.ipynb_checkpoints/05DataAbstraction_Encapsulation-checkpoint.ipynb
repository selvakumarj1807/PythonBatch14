```python
# Create the Class

class LibraryManagement:
    def __init__(self, books):
        self.books = books

    def displayBooks(self):
        for i in self.books:
            print(i, end=', ')

        print('')

    def borrowBook(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"Get your {book} Book Now!...")
        else:
            print('Enter a valid Book Name!...')

    def reciveBook(self, book):
        self.books.append(book)

        print(f"{book} Recieve Book Succesfully!...")
```

```python
## print('Library Management')

msg = '''
1. Display Book
2. Borrow Book
3. Recieve Book
'''

books = ['C', 'C++', 'Java']

libraryObject01 = LibraryManagement(books)

while True:
    print(msg)

    choice = input('Enter your Choice(1/2/3) :')
    
    if choice == '1':
        print('')
        print('Our Library Books')

        libraryObject01.displayBooks()
        
    elif choice == '2':
        book = input('Enter your Borrow Book Name : ')

        libraryObject01.borrowBook(book)

    elif choice == '3':
        book = input('Enter your Recive Book Name : ')

        libraryObject01.reciveBook(book)
        
    else:
        print('Enter a valid choice. Thank You Come Again!...')
        break
```

```python
## print('Library Management')

msg = '''
1. Display Book
2. Borrow Book
3. Recieve Book
'''

books = ['Tamil', 'English', 'Mathematics']

libraryObject02 = LibraryManagement(books)

while True:
    print(msg)

    choice = input('Enter your Choice(1/2/3) :')
    
    if choice == '1':
        print('')
        print('Our Library Books')

        libraryObject02.displayBooks()
        
    elif choice == '2':
        book = input('Enter your Borrow Book Name : ')

        libraryObject02.borrowBook(book)

    elif choice == '3':
        book = input('Enter your Recive Book Name : ')

        libraryObject02.reciveBook(book)
        
    else:
        print('Enter a valid choice. Thank You Come Again!...')
        break
```
