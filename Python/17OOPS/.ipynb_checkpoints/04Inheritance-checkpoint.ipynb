# Inheritance

# Single Inheritance

```python
class Father:
    def playing(self):
        print('Father have Playing Chess')

    def bike(self):
        print('Father have Own Bike')

class Son(Father):
    def reading(self):
        print('Son have Readnig Skill')
```

```python
sonObject01 = Son()

sonObject01.reading()

sonObject01.playing()

sonObject01.bike()
```

# Multiple Inheritance

```python
class Father:
    def playing(self):
        print('Father have Playing Chess')

    def bike(self):
        print('Father have Own Bike')

class Mother:
    def playing(self):
        print('Mother have Playing Chess')

    def cooking(self):
        print('Mother have Cooking Skill')

class Son(Father, Mother):
    def reading(self):
        print('Son have Readnig Skill')
```

```python
sonObject01 = Son()

sonObject01.reading()

sonObject01.playing()

sonObject01.bike()

sonObject01.playing()

sonObject01.cooking()
```

# Multi Level Inheritance

```python
class Father:
    def playing(self):
        print('Father have Playing Chess')

    def bike(self):
        print('Father have Own Bike')

class Mother(Father):
    def playing(self):
        print('Mother have Playing Chess')

    def cooking(self):
        print('Mother have Cooking Skill')

class Son(Mother):
    def reading(self):
        print('Son have Readnig Skill')
```

```python
sonObject01 = Son()

sonObject01.reading()

sonObject01.playing()

sonObject01.bike()

sonObject01.playing()

sonObject01.cooking()
```
