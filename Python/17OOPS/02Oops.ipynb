# Object

```python
# Create the Class

class Student:

    name = 'Priya' 

    age = 27

    def details(self):
        print(self.name, 'your age is', self.age)
```

# Create the Object

```python
studentObject01 = Student()
```

# Call the Class Variable using Object Name

```python
print(studentObject01.name)

print(studentObject01.age)
```

# Call the Class Function Using Object

```python
studentObject01.details()
```

# Values Change Using Object Name

```python
print(studentObject01.name)

print(studentObject01.age)

print('-----------------------------------')

studentObject01.name = 'Dharshini'

studentObject01.age = 30

print(studentObject01.name)

print(studentObject01.age)

print('-----------------------------------')


# Create Another Object

studentObject02 = Student()

print(studentObject02.name)

print(studentObject02.age)

print('-----------------------------------')

studentObject02.name = 'Sam'

studentObject02.age = 22

print(studentObject02.name)

print(studentObject02.age)
```

```python
# Call the Variable From Class

print(Student.name)

print(Student.age)

print('-----------------------------------------------------')

# Call the Variable From Object01

print(studentObject01.name)

print(studentObject01.age)

print('-----------------------------------------------------')

# Call the Variable From Object02

print(studentObject02.name)

print(studentObject02.age)
```

```python
# Call the Class Function Using Object01

studentObject01.details()

# Call the Class Function Using Object02

studentObject02.details()
```

```python
# Create the Class

class Student:
    def __init__(self, name, age):    
        self.name = name 
        self.age = age

    def details(self):
        print(self.name, 'your age is', self.age)
```

```python
# Create the Object 01

studentObject01 = Student('Priya', 27)

studentObject01.details()
```

```python
# Create the Object 02

studentObject02 = Student('Dharshini', 22)

studentObject02.details()
```
