# Create the Class

```python
class Student:
    name = 'Student Name'

    age = 'Student Age'
```

# Call the Class Variable

```python
# Get Attribute Method

print(getattr(Student, 'name'))

# Dot (.) Notation

print(Student.age)
```

# Change The Values

```python
# Dot Notaion

Student.age = 27

print(Student.age)
```

# Create the Class

```python
class User:
    pass
```

# Create the Class Variable. After Class Creation

```python
# Set Attribute method

setattr(User, 'userName', 'admin')

print(User.userName)

# Dot (.) Notation

User.password = 'admin@123'

print(User.password)
```

# Delete The Class Variable. After Class Creation

```python
class Student:
    name = 'Student Name'

    age = 'Student Age'
```

```python
print(Student.name)
```

```python
# dell Attribute

delattr(Student, 'name')

print(Student.name)
```

```python
# Dot (.) Notation

print(Student.age)

del(Student.age)

print(Student.age)
```
