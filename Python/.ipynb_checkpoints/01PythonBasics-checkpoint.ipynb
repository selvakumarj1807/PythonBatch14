# Data Types

1. int(Integer) - 0 to 9
2. float - declare . Values, 10.5,90.6,.....
3. str(String) - declare ''. a to z, A to Z, Special Carecters(`!@#$%^&*,..)
4. bool(Boolean) - True, False

```python
# Variables

name = 'Lara'

age = 25

average = 95.5 

married = True
```

```python
print(name)

print(age)

print(average)

print(married)
```

```python
# To Find the Data Type from the variable

print(type(name))

print(type(age))

print(type(average))

print(type(married))
```

# Concatination

```python
name = 'Lara'

age = 25

print('User Name : ', name)

print('Age : ', age)
```

```python
a = 10

b = 10

print(a + b)

print(a, b)

name = 'Sara'

gender = 'Female'

age = 25

print(name, gender)

print(name, gender, age)

print(name + gender)

print(name + gender , age)
```
