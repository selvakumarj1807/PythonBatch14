# Dictionary
------------

1. Dictionary is a collection of Data type or Built in data types
2. It can declare - { 'Key': 'Value' }
3. It can store n number of Values and Diffrent Data Type Values in a single Variable. Based on Non Index, Baased on Key
4. Dictionary is a mutable(Changeable)
5. Used for Remove Duplicate Values 

```python
userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData, type(userData))
```

# Get Value using Key Name

```python
userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData)

print('----------------------------------')

print(userData['name'])

print('User Name : ', userData['name'])
```

# Add Values using Update method

```python
userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData)

print('------------------------------------------')

v2 = {'mobile': 9500912258, 'city': 'CBE'}

userData.update(v2)

print('------------------------------------------')

print(userData)
```

# Change value using Key Name

```python
userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData)

print('----------------------------------')

userData['name'] = 'Sara'

print(userData)
```

# Remove value using Pop method

```python
userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData)

print('----------------------------------')

userData.pop('gender')

print(userData)
```

# Using For Loop

```python
userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData)

print('----------------------------------')

print(userData.keys())

print(userData.values())

print(userData.items())
```

```python
# For Loop

userData = {'name': 'Dharani', 'age': 25, 'gender': 'Female'}

print(userData)

print('----------------------------------')

for i in userData:
    print(i, ' - ', userData[i])
```

# Dictionary of Dictionary Values

```python
users = {
    'user01': {'name': 'Dharani', 'age': 25, 'gender': 'Female'},
    'user02': {'name': 'Sara', 'age': 22, 'gender': 'Female'},
    'user03': {'name': 'Ram', 'age': 22, 'gender': 'male'}
}

print(users)
```

```python
print(users['user01'])

user01Name = users['user01']['name']

print(user01Name)

user01Age = users['user01']['age']

print(user01Age)
```

```python
users = {
    'user01': {'name': 'Dharani', 'age': 25, 'gender': 'Female'},
    'user02': {'name': 'Sara', 'age': 22, 'gender': 'Female'},
    'user03': {'name': 'Ram', 'age': 23, 'gender': 'male'}
}

print(users)

print('----------------------------------')

for i in users:
    print(i, ':', users[i]['name'], 'your age is',  users[i]['age'])
```
