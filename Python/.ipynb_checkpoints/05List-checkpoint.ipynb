# List
------

1. List is a collection of Data type or Built in data types
2. It can declare - [ ]
3. It can store n number of Values and Diffrent Data Type Values in a single Variable. Based on Index
4. List is a mutable(Changeable)

```python
details = [101, 'Sara', 25, 95.5, True]

print(details, type(details))
```

```python
details = [101, 'Sara', 25, 95.5, True]

print(details[0], type(details[0]))
print(details[1], type(details[1]))
print(details[2], type(details[2]))
print(details[3], type(details[3]))
print(details[4], type(details[4]))

print('------------------------')

print(details[-4], type(details[-4]))
```

```python
details = [101, 'Sara', 25, 95.5, True, [10, 20, 30]]

print(details)

print('--------------------------------------------------')

print(details[5])

print(details[5][1])
```

# List Manipulation

```python
details = [101, 'Sara', 25, 95.5, True, [10, 20, 30]]

print(details)

print('--------------------------------------------------')
# variable[start IndexPosition : End Index Position - 1]

print(details[1:3+1])

print(details[5][1:3+1])

```

# Append

```python
details = [101, 'Sara', 25, 95.5, True]

print(details)

print('----------------------------------------')

details.append(9500912258)

print('----------------------------------------')

print(details)
```

# Extend

```python
details = [101, 'Sara', 25, 95.5, True]

bankDetails = ['PAN0066', '697500654409']

print(details)

print(bankDetails)

print('----------------------------------------')

details.extend(bankDetails)

print('----------------------------------------')

print(details)

print(bankDetails)
```

# Insert

```python
details = [101, 'Sara', 25, 95.5, True]

print(details)

print('----------------------------------------')

details.insert(2, 'Female')

print('----------------------------------------')

print(details)
```

# Change Value Based on Index

```python
details = [101, 'Sara', 25, 95.5, True]

print(details)

print('----------------------------------------')

details[2] = 27

print('----------------------------------------')

print(details)
```

# Remove values Based on Variable's Value

```python
details = [101, 'Sara', 25, 95.5, True]

print(details)

print('----------------------------------------')

details.remove('Sara')

print('----------------------------------------')

print(details)
```

# Remove values Based on Index Position

```python
details = [101, 'Sara', 25, 95.5, True]

print(details)

print('----------------------------------------')

details.pop(1)

print('----------------------------------------')

print(details)
```

# Sort

```python
ageValues = [18, 22, 10, 15, 26, 28, 30, 32, 29, 8, 87, 666, 5, 1502, 10202 ]

print(ageValues)

print('----------------------------------------')

ageValues.sort()

print('----------------------------------------')

print(ageValues)

print('----------------------------------------')

ageValues.sort(reverse=True)

print('----------------------------------------')

print(ageValues)

print('----------------------------------------')

nameValues = ['Sara', 'Ram', 'A', 'sara', 'Br', 'Priya', 'Dharshini']

print('----------------------------------------')

nameValues.sort()

print(nameValues)

print('----------------------------------------')

nameValues.sort(key=len)

print('----------------------------------------')

print(nameValues)
```
