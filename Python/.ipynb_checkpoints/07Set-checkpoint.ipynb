# Set
-----

1. Set is a collection of Data type or Built in data types
2. It can declare - { }
3. It can store n number of Values and Diffrent Data Type Values in a single Variable. Based on Non Index
4. Set is a mutable(Changeable)
5. Used for Remove Duplicate Values 

```python
userData = {10, 'sara', 25, 10, 20, 20, 30}

print(userData, type(userData))
```

# add Value

```python
userData = {1, 2, 3, 4, 5, 6, 1, 2}

print(userData)

print('-------------------------')

userData.add(7)

print('-------------------------')

print(userData)
```

# Update Values

```python
v1 = {1, 2, 3, 4, 5, 1, 2}

v2 = {8, 4, 9, 5, 6, 7, 7}

v1.update(v2)

print(v1)
```

# Remove Value

```python
userData = {1, 2, 3, 4, 5, 6, 1, 2}

print(userData)

print('-------------------------')

userData.remove(2)

print('-------------------------')

print(userData)

print('-------------------------')

userData.remove(9)
```

# Remove Value using discard Method

```python
userData = {1, 2, 3, 4, 5, 6, 1, 2}

print(userData)

print('-------------------------')

userData.discard(2)

print('-------------------------')

print(userData)

print('-------------------------')

userData.discard(9)
```

# Set Operators

```python
v1 = {1, 2, 3, 4, 5, 1, 2, 8}

v2 = {4, 5, 6, 7, 9, 4, 1}

# Union

uV1 = v1.union(v2)

print(v1.union(v2))

print(uV1)
```

```python
v1 = {1, 2, 3, 4, 5, 1, 2, 8}

v2 = {4, 5, 6, 7, 9, 4, 1}

# Intersection

uV1 = v1.intersection(v2)

print(uV1)
```

```python
v1 = {1, 2, 3, 4, 5, 1, 2, 8}

v2 = {4, 5, 6, 7, 9, 4, 1}

# Symmetric_Difference

uV1 = v1.symmetric_difference(v2)

print(uV1)
```

```python
a = {1, 10, 100, 1000, 10000, 100000}

print(a)
```
