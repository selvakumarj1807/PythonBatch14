# While Loop

```python
i = 0 # Value Initial

# 6 <= 5

while i <= 5: # Value Condition
    print(i)

    i = i + 1 # Value Increment
```

```python
i = 0 

while i <= 5: 
    print(i)

    i += 1 
```

# To Print Even Numbers

```python
i = 0 

limit = int(input('Enter your Limit Value : '))

while i <= limit: 
    print(i)

    i += 2 
```

# To Print Odd Numebrs

```python
i = 1 

limit = int(input('Enter your Limit Value : '))

while i <= limit: 
    print(i)

    i += 2 
```

# Jumping Statements

# Break

```python
i = 0 

while i <= 10: 
    print(i)

    if i == 5:
        i += 2
        print('Before Break Statement')
        break
        print('After Break Statement')

    i += 1 
```

# Continue

```python
i = 0 

while i <= 10: 
    print(i)

    if i == 5:
        i += 3
        print('Before continue Statement')
        continue
        print('After continue Statement')

    i += 1 
```

# Pass

```python
i = 0 

while i <= 10: 
    print(i)

    if i == 5:
        i += 3
        print('Before pass Statement')
        pass
        print('After pass Statement')

    i += 1 
```

# While Loop Condition Using Boolean Data Type Values 

```python
i = 0 

while True: 
    print(i)

    if i == 10:
        break

    i += 1 
```

```python
i = 0 

condition = True

while condition: 
    print(i)

    if i == 10:
        condition = False

    i += 1 
```

# Using End

```python
i = 0 

while i <= 5: 
    print(i, end=', ')

    i += 1 
```
