# Python Date Time Module

```python
import datetime
```

```python
currentDate = datetime.date.today()

print(currentDate)
```

# User Date

```python
import datetime as dt

userDate = dt.date(2026, 10, 21) # (year, month, date)

print(userDate)
```

# Date and Time

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)
```

# User Date and Time

```python
userDateTime = dt.datetime(2026, 10, 21, 12, 25, 40)

print(userDateTime)
```

# Split Date Time

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)

print('----------------------------------------')

currentDate = currentDateTime.date()

currentTime = currentDateTime.time()

print('Date : ', currentDate)

print('Time : ', currentTime)
```

# Difference between Two Dates

```python
currentDate = datetime.date.today()

userDate = dt.date(1999, 7, 18) 

diffTwoDates = currentDate - userDate

print(diffTwoDates)
```

# Date Time Format

```python
currentDateTime = dt.datetime.now()

print(currentDateTime)

print('----------------------------------------')

format01 = currentDateTime.strftime('%B')

print(format01)

format02 = currentDateTime.strftime('%d / %b / %Y - %B')

print(format02)
```

# Random Module

```python
import random as rn
```

```python
r1 = rn.randint(1, 10)

print(r1)
```

```python
nameList = ['BA', 'BSc', 'MA', 'MSc']

r2 = rn.choice(nameList)

print(r2)
```
