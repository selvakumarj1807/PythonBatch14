# String Manipulation

```python
msg = 'Hi welcome Python'

print(msg)

print(len(msg))
```

```python
print(msg[0])
```

```python
# variable[starting Position : Ending Position + 1]

print(msg[3:10])

print(msg[11:])

print(msg[:10])

print(msg[-14:-7])
```
