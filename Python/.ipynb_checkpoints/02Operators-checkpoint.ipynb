# Operaotrs:
------------

Some of Operations on Variables or Values.

Types of Operators
------------------

1. Arithmatic Operators
2. ArithmaticAssignment Operators
3. Comparision or Relation Operators
4. Logical Operators
5. Membership Operators
6. Identity Operators
7. Bitwise Operators

```python
# 01 Arithmatic Operators
```

```python
a = 7

b = 2

print('Addition (+) : ', a + b)

print('Subtraction (-) : ', a - b)

print('Multiplication (*) : ', a * b)

print('Division (/) : ', a / b) # Return Quotient value

print('----------------------------------------')

print('Floor Divion (//) : ', a // b)

print('Modulas (%) : ', a % b) # Return Remainder value

print('Exponentiation (**) : ', a ** b) # 7 ** 2 = 7 * 7 = 49
```

# 02 ArithmaticAssignment Operators

```python
a = 10

print(a)

print('-----------------------------')

a += 1 # a = a + 1

print('-----------------------------')

print(a)
```

# 03 Comparision or Relation Operators

<, >, <=, >=, ==, !=

```python
a = 10

b = 5

print(a == b)

print(a != b)

print(a < b) # 10 < 5

print(a > b) # 10 > 5
```

# 04 Logical Operators

and, or, not

```python
userName = input('Enter your Name : ')

password = int(input('Enter your Password : '))

print('--------------------------------------')

print(userName == 'admin')

print(password == 123)

print('--------------------------------------')

print('Logical And : ')

# Logical And
'''
True + True = True
True + False = False
False + True = False
False + False = False
'''

print((userName == 'admin') and (password == 123))

print('--------------------------------------')

print('Logical Or : ')

# Logical Or
'''
True + True = True
True + False = True
False + True = True
False + False = False
'''

print((userName == 'admin') or (password == 123))

print('--------------------------------------')

print('Logical Not : ')

'''
True = False
False = True
'''
print(not(userName == 'admin'))

print(not(((userName == 'admin') or (password == 123))))
```

# 05 Membership Operators

```python
userNames = ['Sara', 'Lara', 'Dharani']

print('Dharani' in userNames)

print('Ram' in userNames)

print('---------------------------------------------')

print('Dharani' not in userNames)

print('Ram' not in userNames)
```

# 06 Identity Operators

```python
a = 10

b = 10

print(id(a))

print(id(b))

print(a is b)

print(a is not b)
```
