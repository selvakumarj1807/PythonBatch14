# Count

```python
msg = 'hi welcome python'

print(msg)

print('-------------------------------------')

countValue = msg.count('o')

print(countValue)
```

# Find

```python
msg = 'hi welcome python'

print(msg)

print('-------------------------------------')

findValue01 = msg.find('o')

print(findValue01)


findValue02 = msg.find('o', 8)

print(findValue02)
```

# Example

```python
gmail = 'priya18@gmail.com'

print(len(gmail))

print(gmail[0:7])

print('-----------------------------------------')

print(gmail.find('@'))

print(gmail[0:gmail.find('@')])
```

# Strip

```python
mobile = '   9500912258   '

print(mobile, type(mobile))

print('--------------------------------')

print(len(mobile))

removeSpace = mobile.strip()

print(len(removeSpace))

print('--------------------------------')

age = 'XX27XX'

print(age)

print('--------------------------------')

print(age.strip('X'))
```

# LStrip

```python
age = 'XX27XX'

print(age)

print('--------------------------------')

print(age.lstrip('X'))
```

# RStrip

```python
age = 'XX27XX'

print(age)

print('--------------------------------')

print(age.rstrip('X'))
```

# Split

```python
userData = "10, Dharani, 27, Female"

print(userData, type(userData))

print('-------------------------------------')

splitUserData = userData.split(', ')

print(splitUserData, type(splitUserData))

print(splitUserData[1])
```

# SplitLines

```python

samplePara = '''Indonesia is named after India because the name is derived from the Greek words Indos (India) and nesos (island),
literally translating to "Indian islands". Coined by 19th-century European academics, 
the term acknowledged the deep, long-standing historical, cultural, 
and trade influences that Indian civilizations had on the Indonesian archipelago. '''

print(samplePara)
```

```python
splitPara = samplePara.splitlines()

print(splitPara)

print('------------------------------------------')

print(splitPara[1])
```

# Join

```python
userData = ['10', 'Dharani', '27', 'Female']

joinData = ' : '.join(userData)

print(joinData, type(joinData))
```

# String Conditions

```python
# isupper()

name = 'ADMIN'

print(name.isupper())
```

```python
# islower()

name = 'admin'

print(name.islower())
```

```python
# isalpha()

name = 'Admin'

print(name.isalpha())
```

```python
# isalnum()

name = 'Admin18'

print(name.isalnum())
```
