# Arbitrary Argument Function

```python
# Create the Function

def userData(*data):
    print(data, type(data))
```

```python
# Call the Function

userData(101, 'Dharani', 27)
```

# Pass List Values

```python
# Create the Function

def userData(data):
    print(data, type(data))
```

```python
# Call the Function

userData([101, 'Dharani', 27])
```

# KeyWord Arguments Function

```python
# Create the Function

def userData(userId, userName, age):
    print(userId, ':', userName,' your age is ', age)
```

```python
# Call the Function

userData(101, 'Dharani', 27)

print('------------------------------------')

userData(userId = 102, age = 22, userName = 'Priya')
```

# Keyword Arbitrary Argument Function

```python
# Create the Function

def userData(**data):
    print(data, type(data))
```

```python
# Call the Function


userData(userId = 101, age = 21, userName = 'Priya')
```

# Default Parametar Function

```python
# Create the Function

def userData(userId, userName, age, state = 'TN'):
    print(userId, ':', userName,' your age is ', age, 'and your state is', state)
```

```python
# Call the Function

userData(101, 'Dharani', 27, 'KL')

print('------------------------------------')

userData(userId = 102, age = 22, userName = 'Priya')
```

# Lambda Function

```python
addTwoNum = lambda a, b: a + b

print('Call the Function In Print Statement : ', addTwoNum(10, 2))

print('----------------------------------------')

answer01 = addTwoNum(20, 4)

print('Call the Function In Variable  : ', answer01)
```

```python
addTwoNum = lambda a, b: f"{a} + {b} = {a + b}"

print('Call the Function In Print Statement : ', addTwoNum(10, 2))

print('----------------------------------------')

answer01 = addTwoNum(20, 4)

print('Call the Function In Variable  : ', answer01)
```
